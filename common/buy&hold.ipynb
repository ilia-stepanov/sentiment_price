#### MFD
начало - max(20200101, дата первой новости, дата первой цены)

конец - min(дата последней новости, дата последней цены)

```python
s = pd.read_csv('/is/prices/mfd_join_lab.csv', sep='|')
s.head(3)
```

```python
p = pd.read_csv('/is/prices/mfd_price.csv', sep='|')
p.date = p.date.astype(float) 
p.ticker = p.ticker.str.lower()
p.head()
```

```python
df_min = p[['date', 'ticker']][p.date>=20200101.0].groupby('ticker').min().rename(columns = {'date':'min_date_price'}).reset_index()
df_min['min_dt'] = df_min.max(axis=1)
df_min = df_min.merge(p[['ticker', 'date', 'close']].rename(columns = {'date':'min_dt'}), how='inner', on=['ticker', 'min_dt'])
df_min.head()
```

```python
df_max = p[['date', 'ticker']].groupby('ticker').max().rename(columns = {'date':'max_date_price'}).reset_index()
df_max['max_dt'] = df_max.min(axis=1)
df_max = df_max.merge(p[['ticker', 'date', 'close']].rename(columns = {'date':'max_dt'}), how='inner', on=['ticker', 'max_dt'])
df_max
```

```python
df_max.merge(df_min, how='inner', on='ticker')
```

```python
mfd = df_max.merge(df_min, how='inner', on='ticker')
mfd = mfd[['ticker', 'min_dt', 'max_dt', 'close_y', 'close_x']]
mfd['bh'] = (mfd.close_x / mfd.close_y - 1) * 100
mfd[['ticker', 'bh']].to_csv('/is/prices/mfd_bh.csv', sep='|', index=False)
mfd.head()
```

#### Magazines
начало - max(дата первой новости, дата первой цены)

конец - min(дата последней новости, дата последней цены)

```python
p = pd.read_csv('/is/prices/mag_join_lab.csv', sep='|')
```

```python
p.head(3)
```

```python
p.ticker = p.ticker.str.lower()
p.head(3)
```

```python
p = p[['ticker', 'date', 'close', 'price_init']]
p.head()
```

```python
df_min = p[['date', 'ticker']].groupby('ticker').min().rename(columns = {'date':'min_date_price'}).reset_index()
df_min['min_dt'] = df_min.max(axis=1)
df_min = df_min.merge(p[['ticker', 'date', 'close']].rename(columns = {'date':'min_dt'}), how='inner', on=['ticker', 'min_dt'])
df_min.head()
```

```python
df_max = p[['date', 'ticker']].groupby('ticker').max().rename(columns = {'date':'max_date_price'}).reset_index()
df_max['max_dt'] = df_max.min(axis=1)
df_max = df_max.merge(p[['ticker', 'date', 'close']].rename(columns = {'date':'max_dt'}), how='inner', on=['ticker', 'max_dt'])
df_max
```

```python
tg = df_max.merge(df_min, how='inner', on='ticker')
tg = tg[['ticker', 'min_dt', 'max_dt', 'close_y', 'close_x']]
tg['bh'] = (tg.close_x / tg.close_y - 1) * 100
tg = tg.drop_duplicates()
```

```python
tg[['ticker', 'bh']].to_csv('/is/prices/mag_bh.csv', sep='|', index=False)
```
